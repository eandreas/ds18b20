```python
def mean(df):
    return df.temp_C.mean()
```

```python
idx = 4
n = 5
delta_mean = 0.06
delta_prev = 0.1

cols = []

for idx in range(0, n):
    cols.append(0)

for idx in range(n, len(df)):

    t_now = df.iloc[idx].temp_C
    t_prev = df.iloc[idx-1].temp_C
    t_mean = mean(df.iloc[idx-(n-1):idx-1])
    
    if t_now > t_prev + delta_prev:
        cols.append(1)
    elif t_now > t_mean + delta_mean:
        cols.append(1)
    elif t_now < t_prev - delta_prev:
        cols.append(-1)
    elif t_now < t_mean - delta_mean:
        cols.append(-1)
    else:
        cols.append(0)
        
len(cols), df.shape
```

```python
fig = go.Figure()


fig.add_trace(go.Scatter(
    x=df.date_time,
    y=df.temp_C,
    line=dict(color='darkgray', width=1, dash='dot')
))

fig.add_trace(go.Scatter(
    x=df.date_time,
    y=df.temp_C,
    marker=dict(
        size=10,
        cmax=1.5,
        cmin=-1.5,
        color=cols,
        colorbar=dict(
            title="Colorbar"
        ),
        colorscale="RdBu"
    ),
    mode="markers")
)

fig.show()
```
