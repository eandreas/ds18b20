```python
import pandas as pd
from pathlib2 import Path
import datetime

temp_file = Path('/Users/eandreas/projects/dev/ds18b20/webapp2/data/get_temp_C.out')
df = pd.read_csv(temp_file, sep=' ', header=None, names=['dev_sn', 'date', 'time', 'temp_raw', 'temp_C'], parse_dates=[['date', 'time']])
df.tail()
```

```python
def npdt2dtdt(npdt):
    return pd.Timestamp(npdt).to_pydatetime()
```

```python
import plotly.graph_objects as go

import pandas as pd
from pathlib2 import Path

# Create figure
fig = go.Figure()

fig.add_trace(
    go.Scatter(x=list(df.date_time), y=list(df.temp_C)))

# Set title
fig.update_layout(
    title_text="Gemessene Temperatur - DS18B20@dragonfly",
    xaxis_title="Datum",
    yaxis_title="Temperatur / C",
    template='none',
    #xaxis_range=[npdt2dtdt(df.date_time.values[-60]), npdt2dtdt(df.date_time.values[-1])]
)

fig.show()
```
